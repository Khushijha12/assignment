Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

ans:- 
LookupError Exception is the Base class for errors raised when something can't be found. 

```python
# key error - A Python KeyError exception is what is raised when you try to access a key that isn’t in a dictionary (dict).
#example
try:
    d={'key': 'khushi' , 'key1' : 'jha' }
    print(d['key2'])
except KeyError as e:
    print(e)
```

```python
# index error = An IndexError means that your code is trying to access an index that is invalid.
# example
try:
    x = [1, 2, 3, 4]
    print(x[5])
except IndexError as e:
    print(e)
```
